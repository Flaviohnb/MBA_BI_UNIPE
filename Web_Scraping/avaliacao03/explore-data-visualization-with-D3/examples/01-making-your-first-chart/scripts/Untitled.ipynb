```python
import pandas as pd
import json
import numpy as np
```

Lendo o arquivo desta forma para tratar os quando houverem registros (dias) com campos ausentes

```python
text=open('./my_weather_data_fahrenheit.json','r')
x=text.read()

y=json.loads(x)
df=pd.DataFrame(y)
```

```python
df.head()
```

```python
df.describe()
```

```python
temperatures = [column for column in df.columns if column.count('emperature') > 0 and column.count('ime') == 0]
temperatures
```

```python
def apply(registro):
    return registro.apply(lambda x: round((x - 32) * 5/9, ndigits=2))
```

```python
df[temperatures].head()
```

```python
df[temperatures] = df[temperatures].apply(apply, axis=1)
```
